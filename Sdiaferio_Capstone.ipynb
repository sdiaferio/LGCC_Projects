Capstone Project - Network Comparison of all NY Providers vs. NY Providers who have opted out of Medicare participation

```python
#Import Pandas for data cleansing and data sets from Github
import pandas as pd
#Upload files from Github
MCROptOut = pd.read_csv('https://raw.githubusercontent.com/sdiaferio/LGCC_Projects/refs/heads/main/Opt_Out_List_March.csv')
NYProviders = pd.read_csv('https://raw.githubusercontent.com/sdiaferio/LGCC_Projects/refs/heads/main/CMS%20NY%20Providers.csv')

```

Review and Cleanse Medicare Opt Out Data

```python
#Review Opt Out Data set to determine if any fields need to be dropped, cleansed, etc.
MCROptOut.head()

```

```python
MCROptOut.info()
```

```python
#Convert Opt Out End Date and Effective Date to Date
MCROptOut['Optout End Date'] = pd.to_datetime(MCROptOut['Optout End Date'])
MCROptOut['Optout Effective Date'] = pd.to_datetime(MCROptOut['Optout Effective Date'])

#Upper case First Name, Last Name, and Specialty
MCROptOut['First Name'] = MCROptOut['First Name'].str.upper()
MCROptOut['Last Name'] = MCROptOut['Last Name'].str.upper()
MCROptOut['Specialty'] = MCROptOut['Specialty'].str.upper()

#Filter data set to only contain NY Providers with an Optout End Date > today
MCROptOutNY = MCROptOut[(MCROptOut['State Code'] == 'NY')& (MCROptOut['Optout End Date'] > 'today')]

#Format Zip code column to only contain 1st 5 characters
MCROptOutNY['Zip Code'] = MCROptOutNY['Zip code'].astype(str).str[:5]

#Only keep needed columns
MCROptOutNY = MCROptOutNY[['npi', 'First Name', 'Last Name', 'Specialty', 'Zip Code','Optout Effective Date', 'Optout End Date']]

#Rename npi column
MCROptOutNY.rename(columns={'npi': 'NPI'}, inplace=True)
MCROptOutNY
```

```python
#Sort data set so most recent effective date is on top for de-duping by NPI
MCROptOutNY = MCROptOutNY.sort_values(by='Optout Effective Date', ascending=False)

#Dedpupe data set
MCROptOutNY = MCROptOutNY.drop_duplicates(subset=['NPI', 'Zip Code'], keep='first')
MCROptOutNY

#Count by NPI and return where the count is greater than 1
NPI_counts = MCROptOutNY['NPI'].value_counts()
NPI_counts[NPI_counts > 1]

#view NPI 1043718240 only to see which column values differ
#MCROptOutNYCheck= MCROptOutNY[MCROptOutNY['NPI'] == 1619158920]
#MCROptOutNYCheck
```

Review and Cleanse NY Provider Data

```python
#Review NY Provider data set to determine if any fields need to be dropped, cleansed, etc.
NYProviders.head()
```

```python
#Format Zip code column to only contain 1st 5 characters
NYProviders['ZIP Code'] = NYProviders['ZIP Code'].astype(str).str[:5]

#Only keep column names that are needed for analysis
NYProviders2 =NYProviders[['NPI', 'Provider Last Name', 'Provider First Name', 'pri_spec', 'ZIP Code']]

#Rename columns to match MCR Provider dataframe headers
NYProviders2.rename(columns={'Provider Last Name': 'Last Name', 'Provider First Name': 'First Name', 'pri_spec': 'Specialty', 'ZIP Code': 'Zip Code'}, inplace=True)
NYProviders2

#Count by NPI and return where the count is greater than 1
NPI_countsNY = NYProviders2['NPI'].value_counts()
NPI_countsNY[NPI_countsNY > 1]

#Check NPI with count more than 1 to see where data is duplicating
NPI_countsNYCheck = NYProviders2[NYProviders2['NPI'] == 1477552933]
NPI_countsNYCheck
```

```python
# Join Opt out and NY provider list on NPI
ProvidersOuter = pd.merge(MCROptOutNY, NYProviders2, on='NPI', how='outer', indicator=True, suffixes=('_MCR', '_NY'))
ProvidersOuter
```

Realign and Rename fields to create one data set for Tableau Visualization

```python
#For Merge on "Both" Keep the NY Fields and rename to remove suffix  _NY
ProvidersBoth = ProvidersOuter[ProvidersOuter['_merge'] == 'both']
ProvidersBoth = ProvidersBoth[['NPI', 'Last Name_NY', 'First Name_NY', 'Specialty_NY', 'Zip Code_NY','Optout Effective Date', 'Optout End Date']]
ProvidersBoth2 = ProvidersBoth.rename(columns={'Last Name_NY': 'Last Name', 'First Name_NY': 'First Name', 'Specialty_NY': 'Specialty', 'Zip Code_NY': 'Zip Code'})
ProvidersBoth2

```

```python
#For Merge on "right only" Keep the NY Fields and rename to remove suffix _NY
ProvidersRight = ProvidersOuter[ProvidersOuter['_merge'] == 'right_only']
ProvidersRight = ProvidersRight[['NPI', 'Last Name_NY', 'First Name_NY', 'Specialty_NY', 'Zip Code_NY', 'Optout Effective Date', 'Optout End Date']]
ProvidersRight2 = ProvidersRight.rename(columns={'Last Name_NY': 'Last Name', 'First Name_NY': 'First Name', 'Specialty_NY': 'Specialty', 'Zip Code_NY': 'Zip Code'})
ProvidersRight2
```

```python
#For Merge on "left only" Keep the MCR Fields and rename to remove suffix _NY
ProvidersLeft= ProvidersOuter[ProvidersOuter['_merge'] == 'left_only']
ProvidersLeft = ProvidersLeft[['NPI', 'Last Name_MCR', 'First Name_MCR', 'Specialty_MCR', 'Zip Code_MCR', 'Optout Effective Date', 'Optout End Date']]
ProvidersLeft2 = ProvidersLeft.rename(columns={'Last Name_MCR': 'Last Name', 'First Name_MCR': 'First Name', 'Specialty_MCR': 'Specialty', 'Zip Code_MCR': 'Zip Code'})
ProvidersLeft2
```

```python
#Union Both, Left, Right tables for final output
ProviderFinal = pd.concat([ProvidersBoth2, ProvidersLeft2, ProvidersRight2])
ProviderFinal
```

```python
#export to an excel to be imported into Tableau for analysis
ProviderFinal.to_excel('ProviderFinal.xlsx', index=False)
```
